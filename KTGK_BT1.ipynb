Bài tập 1: Viết chương trình (hoặc dùng hàm) nhập vào một chuỗi dài chứa nhiều từ. Xuất ra màn hình cùng một chuỗi theo thứ tự từ ngược lại.
Ví dụ:
Input: Tôi là sinh viên CNTT của ĐHKH
Output: ĐHKH của CNTT viên sinh là Tôi

```python
def dao_nguoc_chuoi(input_str):
    tu = input_str.split()  
    tu_dao_nguoc = reversed(tu)  
    chuoi_dao_nguoc = ' '.join(tu_dao_nguoc)  
    return chuoi_dao_nguoc
chuoi_input = input("Nhập chuỗi: ")
ket_qua = dao_nguoc_chuoi(chuoi_input)
print("Chuỗi sau khi đảo ngược: ", ket_qua)
```

Bài tập 2: Điểm KTHP của sinh viên
Điểm môn học của sinh viên bao gồm 2 cột điểm (a,b), a là điểm quá trình, b: điểm thi kết thúc học phần. Điểm môn học của sinh viên được tính như sau: 
diem_mon_hoc = 0.4*a + 0.6*b
Viết các hàm thực hiện các công việc sau:
1. nhap_diem_sv(): Nhập vào một chuỗi n điểm sinh viên từ bàn phím (điểm là một bộ (a,b) với a,b là số thập phân từ 0 đến 10, có 02 giá trị sau dấu phẩy) và trả ra kết quả là một list.
2. diem_mon_học() : lấy danh sách điểm của sinh viên (từ hàm nhap_diem_sv())diem_sv = [(a1,b1),.......,(an,bn)] và trả về kết quả là list chứa điểm môn học của sinh viên: diem_mon_hoc = 0.4*a + 0.6*b
3.diem_lam_tron(): lấy danh sách điểm môn học của sinh viên (từ hàm diem_mon_học()). Hàm này sẽ sử dụng danh sách đầu vào và trả về một danh sách mới với tất cả các điểm được chuyển đổi thành số thập phân với chỉ một giá trị sau dấy phẩy (ví dụ: 9,26 chuyển thành 9.3).
4. dem_sv_failed():  hàm này lấy danh sách điểm môn học của sinh viên sau khi làm tròn và đếm số học sinh không đạt điểm đậu trong danh sách sv_diem. Sinh viên cần có số điểm lớn hơn 5.5 để đạt điểm đậu trong kỳ thi.
5. tren_nguong(): lấy điểm môn học của sinh viên và trả về danh sách tất cả các điểm >= đến ngưỡng.
6. diem_ky_tu (): gán điểm của sinh viên theo các ký tự, trả về kết quả là ký tự tương ứng với điểm của sinh viên
"F" <= 4.0
 4.1 <= "D" <= 5.5
 5.6 <= "C" <= 7.0
 7.1 <= "B" <= 8.5
 8.6 <= "A" <= 10.0
7. ghep_ten_diem(): ghép tên và điểm của sinh viên từ hai danh sách: diem_mon_học và ten_sv. Trả về danh sách thông tin sinh viên chứa tên và điểm của sinh viên ví dụ [["An", 5.6],....,["Nhật", 8.9]]  
8. ds_sv_gioi(): lấy thông tin từ danh sách sinh viên và trả về danh sách các sinh viên có điểm A (từ 8.6 đến 10)

```python
def nhap_diem_sv():
    n = int(input("Nhập số lượng sinh viên: "))
    danh_sach_diem = []
    for i in range(n):
        diem_str = input(f"Nhập điểm của sinh viên {i + 1} (a,b): ")
        a, b = map(float, diem_str.split(","))
        danh_sach_diem.append((a, b))
    return danh_sach_diem

def diem_mon_hoc(diem_sv):
    return [0.4 * a + 0.6 * b for a, b in diem_sv]

def diem_lam_tron(diem_mon_hoc):
    return [round(diem, 1) for diem in diem_mon_hoc]

def dem_sv_failed(diem_sv_rounded):
    return sum(1 for diem in diem_sv_rounded if diem < 5.5)

def tren_nguong(diem_mon_hoc, nguong):
    return [diem for diem in diem_mon_hoc if diem >= nguong]

def diem_ky_tu(diem):
    if diem <= 4.0:
        return "F"
    elif 4.1 <= diem <= 5.5:
        return "D"
    elif 5.6 <= diem <= 7.0:
        return "C"
    elif 7.1 <= diem <= 8.5:
        return "B"
    elif 8.6 <= diem <= 10.0:
        return "A"

def ghep_ten_diem(ten_sv, diem_mon_hoc):
    return list(zip(ten_sv, diem_mon_hoc))

def ds_sv_gioi(ds_sinh_vien):
    return [(ten, diem) for ten, diem in ds_sinh_vien if diem >= 8.6 and diem <= 10]
danh_sach_diem_sv = nhap_diem_sv()
danh_sach_diem_mon_hoc = diem_mon_hoc(danh_sach_diem_sv)
danh_sach_diem_lam_tron = diem_lam_tron(danh_sach_diem_mon_hoc)
so_sinh_vien_truot = dem_sv_failed(danh_sach_diem_lam_tron)
diem_tren_nguong_7 = tren_nguong(danh_sach_diem_mon_hoc, 7)
danh_sach_ky_tu = [diem_ky_tu(diem) for diem in danh_sach_diem_lam_tron]
ten_sv = ["An", "Nhật", "Linh"]
danh_sach_sinh_vien = ghep_ten_diem(ten_sv, danh_sach_diem_lam_tron)
danh_sach_sv_gioi = ds_sv_gioi(danh_sach_sinh_vien)
print("Danh sách điểm môn học:", danh_sach_diem_mon_hoc)
print("Danh sách điểm làm tròn:", danh_sach_diem_lam_tron)
print("Số sinh viên trượt:", so_sinh_vien_truot)
print("Điểm trên ngưỡng 7:", diem_tren_nguong_7)
print("Điểm ký tự:", danh_sach_ky_tu)
print("Danh sách sinh viên:", danh_sach_sinh_vien)
print("Danh sách sinh viên giỏi:", danh_sach_sv_gioi)
```
