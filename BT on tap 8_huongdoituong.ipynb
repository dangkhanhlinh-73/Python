Viết chương trình quản lý sinh viên trong Python. 
Mỗi đối tượng sinh viên có các thuộc tính sau: id, tên, giới tính, tuổi, điểm toán, điểm lý, điểm hóa, điểm trung bình và học lực.

Id là mã sinh viên tự động tăng.

Điểm trung bình là giá trị trung bình của 3 môn toán, lý và hóa.

Học lực được tính như sau:

Giỏi: nếu điểm trung bình lớn hơn hoặc bằng 8.
Khá: nếu điểm trung bình nhỏ hơn 8 và lớn hơn hoặc bằng 6.5.
Trung Bình: nếu điểm trung bình nhỏ hơn 6.5 và lớn hơn hoặc bằng 5.
Yếu: nếu điểm trung bình nhỏ hơn 5.

Hãy xây dựng cá phương thức thực hiện cácchức năng sau:

1. Thêm sinh viên.
2. Cập nhật thông tin sinh viên bởi ID.
3. Xóa sinh viên bởi ID.
4. Tìm kiếm sinh viên theo tên.
5. Sắp xếp sinh viên theo điểm trung bình (GPA).
6. Sắp xếp sinh viên theo tên.
7. Sắp xếp sinh viên theo ID.
8. Hiển thị danh sách sinh viên.

```python
class SinhVien:
    idtieptheo = 1

    def __init__(self, ten, gioitinh, tuoi, diemtoan, diemly, diemhoa):
        self.id = SinhVien.idtieptheo
        SinhVien.idtieptheo += 1
        self.ten = ten
        self.gioitinh = gioitinh
        self.tuoi = tuoi
        self.diemtoan = diemtoan
        self.diemly = diemly
        self.diemhoa = diemhoa
        self.tinhgpa()
        self.tinhhocluc()

    def tinhgpa(self):
        self.gpa = (self.diemtoan + self.diemly + self.diemhoa) / 3

    def tinhhocluc(self):
        if self.gpa >= 8:
            self.hocluc = "Giỏi"
        elif self.gpa >= 6.5:
            self.hocluc = "Khá"
        elif self.gpa >= 5:
            self.hocluc = "Trung Bình"
        else:
            self.hocluc = "Yếu"

class QuanLySinhVien:
    def __init__(self):
        self.danhsachsinhvien = []

    def themsinhvien(self, sinhvien):
        self.danhsachsinhvien.append(sinhvien)

    def capnhatthongtinsinhvien(self, idsinhvien, ten, gioitinh, tuoi, diemtoan, diemly, diemhoa):
        for sinhvien in self.danhsachsinhvien:
            if sinhvien.id == idsinhvien:
                sinhvien.ten = ten
                sinhvien.gioitinh = gioitinh
                sinhvien.tuoi = tuoi
                sinhvien.diemtoan = diemtoan
                sinhvien.diemly = diemly
                sinhvien.diemhoa = diemhoa
                sinhvien.tinhgpa()
                sinhvien.tinhhocluc()
                break

    def xoasinhvien(self, idsinhvien):
        self.danhsachsinhvien = [sv for sv in self.danhsachsinhvien if sv.id != idsinhvien]

    def timkiemsinhvientheoten(self, ten):
        return [sv for sv in self.danhsachsinhvien if ten in sv.ten]

    def sapxepsinhvientheogpa(self):
        self.danhsachsinhvien.sort(key=lambda sv: sv.gpa, reverse=True)

    def sapxepsinhvientheoten(self):
        self.danhsachsinhvien.sort(key=lambda sv: sv.ten)

    def sapxepsinhvientheoid(self):
        self.danhsachsinhvien.sort(key=lambda sv: sv.id)

    def hienthidanhsachsinhvien(self):
        for sinhvien in self.danhsachsinhvien:
            print(f"ID: {sinhvien.id}, Tên: {sinhvien.ten}, Giới tính: {sinhvien.gioitinh}, Tuổi: {sinhvien.tuoi}, GPA: {sinhvien.gpa}, Học lực: {sinhvien.hocluc}")
if __name__ == "__main__":
    quanly = QuanLySinhVien()

    while True:
        print("\nMenu:")
        print("1. Thêm sinh viên")
        print("2. Cập nhật thông tin sinh viên")
        print("3. Xóa sinh viên")
        print("4. Tìm kiếm sinh viên theo tên")
        print("5. Sắp xếp sinh viên theo GPA")
        print("6. Sắp xếp sinh viên theo tên")
        print("7. Sắp xếp sinh viên theo ID")
        print("8. Hiển thị danh sách sinh viên")
        print("9. Thoát")
        chon = input("Chọn tùy chọn: ")

        if chon == "1":
            ten = input("Tên: ")
            gioitinh = input("Giới tính: ")
            tuoi = int(input("Tuổi: "))
            diemtoan = float(input("Điểm Toán: "))
            diemly = float(input("Điểm Lý: "))
            diemhoa = float(input("Điểm Hóa: "))
            sinhvien = SinhVien(ten, gioitinh, tuoi, diemtoan, diemly, diemhoa)
            quanly.themsinhvien(sinhvien)
            print("Sinh viên đã được thêm vào danh sách.")

        elif chon == "2":
            idsinhvien = int(input("Nhập ID sinh viên cần cập nhật: "))
            ten = input("Tên mới: ")
            gioitinh = input("Giới tính mới: ")
            tuoi = int(input("Tuổi mới: "))
            diemtoan = float(input("Điểm Toán mới: "))
            diemly = float(input("Điểm Lý mới: "))
            diemhoa = float(input("Điểm Hóa mới: "))
            quanly.capnhatthongtinsinhvien(idsinhvien, ten, gioitinh, tuoi, diemtoan, diemly, diemhoa)
            print("Thông tin sinh viên đã được cập nhật.")

        elif chon == "3":
            idsinhvien = int(input("Nhập ID sinh viên cần xóa: "))
            quanly.xoasinhvien(idsinhvien)
            print("Sinh viên đã bị xóa khỏi danh sách.")

        elif chon == "4":
            ten = input("Nhập tên sinh viên cần tìm kiếm: ")
            sinhvientimthay = quanly.timkiemsinhvientheoten(ten)
            if sinhvientimthay:
                print("Kết quả tìm kiếm:")
                for sv in sinhvientimthay:
                    print(f"ID: {sv.id}, Tên: {sv.ten}, GPA: {sv.gpa}, Học lực: {sv.hocluc}")
            else:
                print("Không tìm thấy sinh viên có tên này.")

        elif chon == "5":
            quanly.sapxepsinhvientheogpa()
            print("Danh sách sinh viên đã được sắp xếp theo GPA.")
            quanly.hienthidanhsachsinhvien()

        elif chon == "6":
            quanly.sapxepsinhvientheoten()
            print("Danh sách sinh viên đã được sắp xếp theo tên.")
            quanly.hienthidanhsachsinhvien()

        elif chon == "7":
            quanly.sapxepsinhvientheoid()
            print("Danh sách sinh viên đã được sắp xếp theo ID.")
            quanly.hienthidanhsachsinhvien()

        elif chon == "8":
            quanly.hienthidanhsachsinhvien()

        elif chon == "9":
            break

        else:
            print("Tùy chọn không hợp lệ. Vui lòng chọn lại.")
```
