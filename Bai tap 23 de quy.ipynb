Bài tập 1: Viết chương trình Python đệ quy để tính tổng của một danh sách các số

```python
def tinh_tong_ds_recursive(danh_sach):
    # Kiểm tra điều kiện dừng
    if len(danh_sach) == 0:
        return 0
    else:
        # Gọi đệ quy với danh sách bỏ đi phần tử đầu tiên
        return danh_sach[0] + tinh_tong_ds_recursive(danh_sach[1:])
danh_sach_so = [1, 2, 3, 4, 5]
ket_qua = tinh_tong_ds_recursive(danh_sach_so)
print(f"Tổng của danh sách {danh_sach_so} là: {ket_qua}")
```

Bài tập 2: Viết chương trình Python đệ quy tính tổng danh sách
Dữ liệu test: [1,2, [3,4],[5,6],[7,[8,9]]]
Kết quả mong muốn: 45

```python
def tinh_tong(lst):
    tong = 0
    for phan_tu in lst:
        if isinstance(phan_tu, list):
            tong += tinh_tong(phan_tu)
        else:
            tong += phan_tu
    return tong

# Dữ liệu test
du_lieu_test = [1, 2, [3, 4], [5, 6], [7, [8, 9]]]
ket_qua = tinh_tong(du_lieu_test)
print("Kết quả mong muốn:", ket_qua)
```

Bài tập 3: Viết chương trình Python lấy giai thừa của một số nguyên không âm

```python
def tinh_giai_thua(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * tinh_giai_thua(n - 1)
so_nguyen = int(input("Nhập một số nguyên không âm: "))

if so_nguyen < 0:
    print("Vui lòng nhập một số nguyên không âm.")
else:
    ket_qua = tinh_giai_thua(so_nguyen)
    print(f"Giai thừa của {so_nguyen} là: {ket_qua}")
```

Bài tập 4: Viết chương trình Python giải dãy Fibonacci bằng đệ quy

```python
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)
nterms = int(input("Nhập số lượng số Fibonacci muốn hiển thị: "))
if nterms <= 0:
    print("Vui lòng nhập một số nguyên dương.")
else:
    print("Dãy Fibonacci:")
    for i in range(nterms):
        print(fibonacci(i))
```

Bài tập 5: Viết chương trình Python để lấy tổng các chữ số của một số nguyên không âm bằng đệ quy
Dữ liệu kiểm tra:
sumDigits(23456) -> 20
sumDigits(345) -> 12

```python
def sumDigits(n):
    # Base case: n là số có một chữ số
    if n < 10:
        return n
    else:
        
        return n % 10 + sumDigits(n // 10)
print(sumDigits(23456)) 
print(sumDigits(345))   
```

Bài tập 6: Viết chương trình Python để tính tổng các số nguyên dương của n+(n-2)+(n-4)... (until n-x =< 0)
dữ liệu kiểm tra:
tong_day(6) -> 12
tong_day(10) -> 30

```python
def tong_day(n):
    # Kiểm tra nếu n là số lẻ, giảm n xuống một để làm cho n là số chẵn
    if n % 2 != 0:
        n -= 1

    # Khởi tạo biến tổng
    total = 0

    # Tính tổng các số nguyên dương theo công thức đã cho
    for i in range(n, 0, -2):
        total += i

    return total

# Kiểm tra với dữ liệu mẫu
print(tong_day(6))  # Kết quả: 12
print(tong_day(10)) # Kết quả: 30
```

Bài tập 7: Viết chương trình Python để tính tổng điều hòa của n-1.
Tổng điều hòa là tổng nghịch đảo của các số nguyên dương..
ví dụ: 1+1/2+1/3+1/4+1/5+....

```python
def tong_dieu_hoa(n):
    if n <= 0:
        print("Vui lòng nhập một số nguyên dương.")
        return

    tong = 0
    for i in range(1, n):
        tong += 1/i

    return tong
n = 5
ket_qua = tong_dieu_hoa(n)

print(f"Tổng điều hòa của {n-1} số là: {ket_qua:.4f}")
```

Bài tập 8: Viết chương trình Python để tính giá trị của 'a' lũy thừa 'b' bằng đệ quy
Dữ liệu test
luythua(3,4) -> 81

```python
def luythua(a, b):
    # Trường hợp cơ bản: a^0 = 1
    if b == 0:
        return 1
    # Trường hợp đệ quy: a^b = a * a^(b-1)
    else:
        return a * luythua(a, b-1)
result = luythua(3, 4)
print(result)
```

Bài tập 9: Viết chương trình Python tìm ước chung lớn nhất (gcd) của hai số nguyên bằng đệ quy.

```python
def gcd_recursive(a, b):
    if b == 0:
        return a
    else:
        return gcd_recursive(b, a % b)
num1 = int(input("Nhập số nguyên thứ nhất: "))
num2 = int(input("Nhập số nguyên thứ hai: "))
result = gcd_recursive(num1, num2)
print(f"Ước chung lớn nhất của {num1} và {num2} là: {result}")
```
