Câu 3. (Kiểm tra số nguyên tố) Cho 1 số tự nhiên n (n<=10^12) Kiểm tra xem n có phải là 1 số nguyên tố hay không?

```python
import math

def is_primese(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

n = int(input("Nhập số tự nhiên n: "))

if is_primese(n):
    print(f"{n} là số nguyên tố")
else:
    print(f"{n} không là số nguyên tố")
```
