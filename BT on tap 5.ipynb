Bài tập 1. Viết chương trình Python để thực hiện các thao tác sau:
1. Nhập vào 01 mật khẩu từ người dùng và kiểm tra mật khẩu đó có hợp lệ không: 

Các tiêu chí kiểm tra mật khẩu hợp lệ bao gồm:
i. Độ dài mật khẩu tối thiểu: 6
ii. Độ dài mật khẩu tối đa: 14
iii. Chứa ít nhất một cái
iv. Chứa ít nhất một số
v. Chứa ít nhất một chữ cái in hoa

2. Viết hàm với đầu vào là một chuỗi mật khẩu phân tách nhau bởi dấu phẩy và đầu ra là danh sách các mật khẩu hợp lệ.

```python
import re
def la_mat_khau_hop_le(mat_khau):
    if len(mat_khau) < 6:
        return False
    if len(mat_khau) > 14:
        return False
    if not any(char.isalpha() for char in mat_khau):
        return False
    if not any(char.isdigit() for char in mat_khau):
        return False
    if not any(char.isupper() for char in mat_khau):
        return False
    return True

def main():
    danh_sach_mat_khau_input = input("Nhập danh sách mật khẩu (phân tách bởi dấu phẩy): ")
    danh_sach_mat_khau = danh_sach_mat_khau_input.split(',')
    mat_khau_hop_le = [mat_khau for mat_khau in danh_sach_mat_khau if la_mat_khau_hop_le(mat_khau)]
    if mat_khau_hop_le:
        print("Các mật khẩu hợp lệ là:")
        for mat_khau in mat_khau_hop_le:
            print(mat_khau)
    else:
        print("Không có mật khẩu hợp lệ.")

if __name__ == "__main__":
    main()
```
