Bài tập 1. Viết một hàm Python chấp nhận một chuỗi và tính số lượng chữ hoa và chữ thường.

```python
def dem_chu_hoa_va_chu_thuong(chuoi):
    dem_chu_hoa = 0
    dem_chu_thuong = 0
    for ky_tu in chuoi:
        if ky_tu.isupper():
            dem_chu_hoa += 1
        elif ky_tu.islower():
            dem_chu_thuong += 1
    return dem_chu_hoa, dem_chu_thuong
chuoi_test = " Việt Nam"
so_chu_hoa, so_chu_thuong = dem_chu_hoa_va_chu_thuong(chuoi_test)
print(f"Số chữ hoa: {so_chu_hoa}")
print(f"Số chữ thường: {so_chu_thuong}")
```

Bài tập 2. Viết hàm liệt kê số nguyên tố trong đoạn từ A đến B. (a,b là số tự nhiên, a<=b)

```python
def la_so_nguyen_to(num):
    """Kiểm tra xem một số có phải là số nguyên tố hay không."""
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def liet_ke_so_nguyen_to_trong_doan(a, b):
    """Liệt kê tất cả các số nguyên tố trong đoạn từ a đến b."""
    so_nguyen_to = [num for num in range(a, b + 1) if la_so_nguyen_to(num)]
    return so_nguyen_to
a = 10
b = 50
ket_qua = liet_ke_so_nguyen_to_trong_doan(a, b)
print(f"Các số nguyên tố trong đoạn từ {a} đến {b}: {ket_qua}")
```

Bài tập 3: Viết hàm trả về chuỗi con đảo ngược từ vị trí a đến b của chuỗi S

```python
def dao_nguoc_chuoi_con(s, a, b):
    if a < 0 or b >= len(s) or a > b:
        return "Chỉ số không hợp lệ"
    
    chuoi_con_can_dao = s[a:b + 1]
    chuoi_con_dao_nguoc = chuoi_con_can_dao[::-1]
    
    ket_qua = s[:a] + chuoi_con_dao_nguoc + s[b + 1:]
    
    return ket_qua
S = "thế giới"
a = 3
b = 7
ket_qua = dao_nguoc_chuoi_con(S, a, b)
print(ket_qua)
```

Bài tập 4. Viết hàm sắp xếp danh sách số thực theo thứ tự tăng dần. (Không sử dụng hàm sắp xếp có sẵn).

```python
def sap_xep_tang_dan(danh_sach):
    n = len(danh_sach)
    for i in range(n):
        for j in range(0, n-i-1):
            if danh_sach[j] > danh_sach[j+1]:
                danh_sach[j], danh_sach[j+1] = danh_sach[j+1], danh_sach[j]
danh_sach_so_thuc = [5.4, 3.2, 8.1, 2.7, 1.6]
print("Danh sách trước khi sắp xếp:", danh_sach_so_thuc)

sap_xep_tang_dan(danh_sach_so_thuc)

print("Danh sách sau khi sắp xếp:", danh_sach_so_thuc)
```

Bài tập 5. Viết hàm trả về hai danh sách sau khi đã hoán đổi nữa sau danh sách cho nhau. (Tham số là 2 danh sách bất kỳ).

```python
def hoan_doi_nua_sau(danh_sach1, danh_sach2):
    # Kiểm tra nếu danh sách không đủ dài để hoán đổi
    if len(danh_sach1) < 2 or len(danh_sach2) < 2:
        print("Danh sách không đủ dài để hoán đổi nữa sau.")
        return None
    nua_sau_danh_sach1 = len(danh_sach1) // 2
    nua_sau_danh_sach2 = len(danh_sach2) // 2
    danh_sach1[:nua_sau_danh_sach1], danh_sach2[:nua_sau_danh_sach2] = danh_sach2[:nua_sau_danh_sach2], danh_sach1[:nua_sau_danh_sach1]

    return danh_sach1, danh_sach2
danh_sach_a = [1, 2, 3, 4, 5]
danh_sach_b = [6, 7, 8, 9, 10]

ket_qua = hoan_doi_nua_sau(danh_sach_a, danh_sach_b)
print("Danh sách sau khi hoán đổi nửa sau:")
print("Danh sách A:", ket_qua[0])
print("Danh sách B:", ket_qua[1])
```
