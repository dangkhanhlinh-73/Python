Câu 1. (2.0 điểm): 
-Viết hàm trả về từ cuối cùng của một chuỗi cho trước
Ví dụ:
    Input: Nguyễn Văn Anh
    Output: Anh
-Nhập vào danh sách họ tên của các sinh viên, quá trình nhập dừng khi họ tên nhập vào bằng rỗng. In ra màn hình số lượng sinh viên có tên là “Anh”.

```python
def tucuoi(s):
    a=s.split(' ')
    return a[len(a)-1]
a=[]
d=0
while True:
    s=input()
    if s=='':
        break
    else:
        a.append(s)
d=0
for s in a:
    if(tucuoi(s)=='Anh'):
        d=d+1
print('Số lượng sinh viên có tên Anh là: ',d)
```

Câu 2. (2.0 điểm): Thông tin mỗi học phần bao gồm: mã học phần (mahp), tên học phần (tenhp), số tín chỉ (sotc). Sử dụng kiểu từ điển bao gồm các cặp key-value để biểu diễn các thông tin cho mỗi học phần như ví dụ: {“mahp”: “TIN1103”, “tenhp”, “Lập trình Python”, “sotc”:3}.
Viết chương trình thực hiện các yêu cầu sau:
-	Nhập vào danh sách có n (n<10) học phần.
-	In ra màn hình thông tin của các học phần thuộc ngành CNTT (mahp có 3 ký tự đầu là “TIN”).
-	Nhập vào một chuỗi s, in ra màn hình các học phần có tên học phần chứa chuỗi s. Ví dụ: 		Input: s= “lập trình”
Output: “Lập trình nâng cao”, “Lập trình Python”,… 

```python
n=int(input('Nhập số học phần: '))
DS_Hoc_phan=[]
for i in range(n):
    ma=input('Nhập mã học phần: ')
    ten=input('Nhập tên học phần: ')
    tc=int(input('Nhập số tín chỉ: '))
    a={}
    a['mahp']=ma
    a['tenhp']=ten
    a['sotc']=tc
    DS_Hoc_phan.append(a)
print(DS_Hoc_phan)
```

Câu 3. (3.0 điểm):
Xây dựng lớp HOADON mô tả thông tin hóa đơn sử dụng điện của khách hàng, bao gồm các thuộc tính sau: mã khách hàng, họ tên, địa chỉ, chỉ số cũ, chỉ số mới. Viết các phương thức cho lớp HOADON bao gồm:
- Phương thức khởi tạo;
- Phương thức trả về mã khách hàng;
- Phương thức trả về số lượng điện sử dụng theo công thức: lượng điện tiêu thụ= chỉ số mới- chỉ số cũ.
- Phương thức hiển thị thông tin của các thuộc tính mã khách hàng, họ tên, địa chỉ, số lượng điện tiêu thụ.
- Toán tử tải bội > để so sánh hai đối tượng lớp HOADON dựa vào tiêu chí lượng điện tiêu thụ.

```python
class hoadon:
    def __init__(self,ma,hoten,diachi,csc,csm):
        self.makh=ma
        self.hoten=hoten
        self.diachi=diachi
        self.csc=csc
        self.csm=csm
    def getmakh(self):
        return self.makh
    def gethoten(self):
        return self.hoten
    def getdiachi(self):
        return self.diachi
    def LuongDienTieuThu(self):
        return self.csm-self.csc
    def show(self):
        txt="{:<15} {:<30} {:<30} {:<10} {:<10} {:<10}"
        print(txt.format(self.makh,self.hoten,self.diachi,self.csc,self.csm,self.LuongDienTieuThu()))
    def __gt__(self,other):
        return self.LuongDienTieuThu()>other.LuongDienTieuThu()
```

- Lưu thông tin của những khách hàng (gồm mã khách hàng, họ tên, địa chỉ) có số lượng điện tiêu thụ >100 vào file văn bản sao cho thông tin mỗi khách hàng được lưu trên một dòng, trong đó các thuộc tính được viết cách nhau bằng dấu chấm phẩy (;).
Ví dụ: KH0001; Nguyen Van Anh; 17 Nguyen Hue   

```python
n=int(input('Nhập n: '))
Khachhang=[]
Ma=[]
for i in range(n):
    ma=input('Nhập mã khách hàng:')
    while(ma in Ma):
        ma=input('Nhập mã khách hàng:')
    Ma.append(ma)
    hoten=input('Nhập họ tên khách hàng:')
    diachi=input('Nhập địa chỉ khách hàng:')
    csc=int(input('Nhập chỉ số cũ: '))
    csm=int(input('Nhập chỉ số mới: '))
    a=hoadon(ma,hoten,diachi,csc,csm)
    Khachhang.append(a)
for i in range(n):
    Khachhang[i].show()
```
