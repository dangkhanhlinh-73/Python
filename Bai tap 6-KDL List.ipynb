Câu 1. (Tổng các số nguyên lẻ) Nhập vào số nguyên dương n và DS gồm n số nguyên a_0,a_1,...,a_(n-1). Hãy tính tổng các số nguyên lẻ trong dãy.


```python
# Nhập số nguyên dương n
n = int(input("Nhập số nguyên dương n: "))

# Khởi tạo biến tổng
tongle = 0

# Nhập danh sách số nguyên a_i và tính tổng các số lẻ
for i in range(n):
    a = int(input(f"Nhập số nguyên thứ {i + 1}: "))
    if a % 2 == 1:
        tongle += a

# In kết quả
print(f"Tổng các số nguyên lẻ trong dãy là: {tongle}")












```

```python
# Nhập số nguyên dương n
n = int(input("Nhập số nguyên dương n: "))

# Khởi tạo danh sách số nguyên
ds = []

# Nhập danh sách số nguyên a_i
for i in range(n):
    a = int(input(f"Nhập số nguyên thứ {i + 1}: "))
    ds.append(a)

# Sử dụng hàm list để lọc ra các số nguyên lẻ
sole = [x for x in ds if x % 2 == 1]

# Tính tổng các số lẻ bằng hàm sum
tongle = sum(sole)

# In kết quả
print(f"Tổng các số nguyên lẻ trong dãy là: {tongle}")
```

Câu 2. (Tổng k số hạng liên tiếp có tổng lớn nhất) NHập vào số nguyên dương n, k (k<=n), và DS gồm n số nguyên a_0,a_1,...,a_(n-1). Tìm k phần tử liên tiếp trong dãy: a_i,a_(i+1), ..., a_(i+k-1) sao cho tổng a_i + a_(i+1) + ...+ a_(i+k-1) đạt GTLN. Ví dụ: dãy gồm 6 số hạng: 1,2,-19,3,4,-1 với k=3 có tổng 3 số hạng liên tiếp lớn nhất là: 3,4,-1 với tổng 3+4-1=6.

```python
def tim_tong_lon_nhat_list(n, k, arr):
    tong_lon_nhat = float('-inf')
    day_con_tong_lon_nhat = []

    for i in range(n - k + 1):
        day_con = arr[i:i+k]
        tong_hien_tai = sum(day_con)

        if tong_hien_tai > tong_lon_nhat:
            tong_lon_nhat = tong_hien_tai
            day_con_tong_lon_nhat = day_con

    return day_con_tong_lon_nhat

# Nhập dữ liệu
n = int(input("Nhập số lượng phần tử của dãy: "))
k = int(input("Nhập k (số phần tử liên tiếp): "))
arr = []

for i in range(n):
    arr.append(int(input(f"Nhập phần tử thứ {i+1}: ")))

ket_qua = tim_tong_lon_nhat_list(n, k, arr)
print(f"Tổng {k} số hạng liên tiếp lớn nhất là: {ket_qua} với tổng {sum(ket_qua)}")
```

```python
def tim_tong_lon_nhat(n, k, arr):
    tong_lon_nhat = float('-inf')
    day_con_tong_lon_nhat = []

    for i in range(n - k + 1):
        tong_hien_tai = 0
        for j in range(i, i+k):
            tong_hien_tai += arr[j]

        if tong_hien_tai > tong_lon_nhat:
            tong_lon_nhat = tong_hien_tai
            day_con_tong_lon_nhat = arr[i:i+k]

    return day_con_tong_lon_nhat

# Nhập dữ liệu
n = int(input("Nhập số lượng phần tử của dãy: "))
k = int(input("Nhập k (số phần tử liên tiếp): "))
arr = []

for i in range(n):
    arr.append(int(input(f"Nhập phần tử thứ {i+1}: ")))

ket_qua = tim_tong_lon_nhat(n, k, arr)
print(f"Tổng {k} số hạng liên tiếp lớn nhất là: {ket_qua} với tổng {sum(ket_qua)}")
```
