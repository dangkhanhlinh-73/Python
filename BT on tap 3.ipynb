Bài tập 1: Viết một chương trình Python thực hiện các thao tác sau:
1.Nhập hai danh sách DS1, DS2 các số nguyên có cùng độ dài
2.Lọc các số lẻ từ danh sách DS1 dùng hàm filter
3.Dùng map() và lambda để tạo danh sách chứa các giá trị bình phương của các số trong DS1 sau đó in một nửa số giá trị đầu tiên của danh sách trong 1 dòng và 1 nửa số giá trị cuối trong 1 dòng.
4.Sắp xếp các phần tử trong DS1 theo thứ tự giảm dần
5.Sắp xếp các phần tử trong DS1 theo khoảng cách đến số k cho trước 
6.Tạo một hàm có thể tạo ra một dictionary chứa key là những số từ DS1 và các giá trị bình phương của key 
7.Tạo danh sách mới mà các phần tử trong list là tổng của các phần tử tương ứng trong DS1 và DS2 
8.Tạo danh sách mới mà các phần tử trong list là True nếu phần tử tương ứng trong DS1 lớn hơn phần tử tương ứng trong DS2 và trả về False nếu ngược lại

```python
# Nhập hai danh sách DS1, DS2
DS1 = list(map(int, input("Nhập danh sách DS1 (các số nguyên, cách nhau bằng dấu cách): ").split()))
DS2 = list(map(int, input("Nhập danh sách DS2 (các số nguyên, cách nhau bằng dấu cách): ").split()))

# 2. Lọc các số lẻ từ danh sách DS1
DS1_le = list(filter(lambda x: x % 2 != 0, DS1))
print("Các số lẻ từ DS1:", DS1_le)

# 3. Dùng map() và lambda để tạo danh sách chứa các giá trị bình phương của các số trong DS1
DS1_binh_phuong = list(map(lambda x: x**2, DS1))
nua_dau = DS1_binh_phuong[:len(DS1_binh_phuong)//2]
nua_cuoi = DS1_binh_phuong[len(DS1_binh_phuong)//2:]
print("Nửa đầu của danh sách bình phương:", nua_dau)
print("Nửa cuối của danh sách bình phương:", nua_cuoi)

# 4. Sắp xếp các phần tử trong DS1 theo thứ tự giảm dần
DS1_giam_dan = sorted(DS1, reverse=True)
print("DS1 sau khi sắp xếp giảm dần:", DS1_giam_dan)

# 5. Sắp xếp các phần tử trong DS1 theo khoảng cách đến số k cho trước
k = int(input("Nhập số k: "))
DS1_khoang_cach = sorted(DS1, key=lambda x: abs(x - k))
print("DS1 sau khi sắp xếp theo khoảng cách đến", k, ":", DS1_khoang_cach)

# 6. Tạo một hàm có thể tạo ra một dictionary chứa key là những số từ DS1 và các giá trị bình phương của key
def tao_dictionary(ds):
    return {x: x**2 for x in ds}

DS1_dictionary = tao_dictionary(DS1)
print("Dictionary từ DS1:", DS1_dictionary)

# 7. Tạo danh sách mới mà các phần tử trong list là tổng của các phần tử tương ứng trong DS1 và DS2
DS_tong = [x + y for x, y in zip(DS1, DS2)]
print("Danh sách tổng của DS1 và DS2:", DS_tong)

# 8. Tạo danh sách mới mà các phần tử trong list là True nếu phần tử tương ứng trong DS1 lớn hơn phần tử tương ứng trong DS2 và trả về False nếu ngược lại
DS_lon_hon = [x > y for x, y in zip(DS1, DS2)]
print("Danh sách các phần tử trong DS1 lớn hơn phần tử tương ứng trong DS2:", DS_lon_hon)
```
