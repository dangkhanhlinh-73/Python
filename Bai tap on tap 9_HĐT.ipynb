    + Điểm Toán (Số thực, không âm)

-	Có phương thức khởi tạo

    + Khá: nếu điểm trung bình nhỏ hơn 8 và lớn hơn hoặc bằng 6.5, điểm môn Toán >=5.0, điểm môn văn  >=5.0,
    + Trung Bình: nếu điểm trung bình nhỏ hơn 6.5 và lớn hơn hoặc bằng 5.
    + Yếu: nếu điểm trung bình nhỏ hơn 5.",
Viết chương trình Python cho phép khai báo 01 danh sách gồm n học sinh thuộc lớp HOC_SINH. Thực hiện các công việc sau: 
a.	Sử dụng phương thức nhập và hiển thị dữ liệu để nhập dữ liệu tương ứng cho các đối tượng này và hiển thị các đối tượng ra màn hình.
b.	Tìm sinh viên có điểm trung bình lớn nhất và nhỏ nhất.
c.	Sắp xếp danh sách sinh viên theo thứ tự tăng dần theo điểm môn Toán.
d.	Sắp xếp danh sách sinh viên theo tiêu chí điểm trung bình gần 7.0.
e.	Liệt kê các học sinh có năm sinh nhỏ hơn 2014 trong danh sách và xuất ra màn hình họ tên sinh viên.
f.	Hãy liệt kê tất cả các sinh viên có năm sinh từ 2012 đến 2015 và có học lực giỏi để lập danh sách khen thưởng và in ra màn hình họ tên sinh viên.


```python
class HOC_SINH:
    def __init__(self, ma_so, ho_ten, nam_sinh, diem_toan, diem_van):
        self.ma_so = ma_so
        self.ho_ten = ho_ten
        self.nam_sinh = nam_sinh
        self.diem_toan = diem_toan
        self.diem_van = diem_van

    def nhap_thong_tin(self):
        self.ma_so = int(input("Nhập mã số học sinh: "))
        self.ho_ten = input("Nhập họ tên học sinh: ")
        self.nam_sinh = int(input("Nhập năm sinh: "))
        self.diem_toan = float(input("Nhập điểm Toán: "))
        self.diem_van = float(input("Nhập điểm Văn: "))


    def hien_thi_thong_tin(self):
        print(f"Mã số học sinh: {self.ma_so}")
        print(f"Họ tên: {self.ho_ten}")
        print(f"Năm sinh: {self.nam_sinh}")
        print(f"Điểm Toán: {self.diem_toan}")
        print(f"Điểm Văn: {self.diem_van}")

    def diem_trung_binh(self):
        return (self.diem_toan + self.diem_van) / 2

    def xep_loai(self):
        if self.diem_trung_binh() >= 8 and self.diem_toan >= 6.5 and self.diem_van >= 6.5:
            return "Giỏi"
        elif 6.5 <= self.diem_trung_binh() < 8 and self.diem_toan >= 5.0 and self.diem_van >= 5.0:
            return "Khá"
        elif 5.0 <= self.diem_trung_binh() < 6.5:
            return "Trung Bình"
        else:
            return "Yếu"

n = int(input("Nhập số lượng học sinh: "))
danh_sach_hoc_sinh = []

for i in range(n):
    print(f"Nhập thông tin cho học sinh thứ {i + 1}:")
    hoc_sinh = HOC_SINH(0, "", 0, 0, 0)
    hoc_sinh.nhap_thong_tin()
    danh_sach_hoc_sinh.append(hoc_sinh)
print("Thông tin các học sinh:")
for hoc_sinh in danh_sach_hoc_sinh:
    hoc_sinh.hien_thi_thong_tin()
diem_trung_binh_max = max(danh_sach_hoc_sinh, key=lambda hs: hs.diem_trung_binh())
diem_trung_binh_min = min(danh_sach_hoc_sinh, key=lambda hs: hs.diem_trung_binh())
print("Học sinh có điểm trung bình cao nhất:")
diem_trung_binh_max.hien_thi_thong_tin()
print("Học sinh có điểm trung bình thấp nhất:")
diem_trung_binh_min.hien_thi_thong_tin()
danh_sach_hoc_sinh.sort(key=lambda hs: hs.diem_toan)
print("Danh sách học sinh theo điểm môn Toán tăng dần:")
for hoc_sinh in danh_sach_hoc_sinh:
    hoc_sinh.hien_thi_thong_tin()
danh_sach_hoc_sinh.sort(key=lambda hs: abs(hs.diem_trung_binh() - 7.0))
print("Danh sách học sinh theo điểm trung bình gần 7.0:")
for hoc_sinh in danh_sach_hoc_sinh:
    hoc_sinh.hien_thi_thong_tin()
print("Học sinh có năm sinh nhỏ hơn 2014:")
for hoc_sinh in danh_sach_hoc_sinh:
    if hoc_sinh.nam_sinh < 2014:
        print(hoc_sinh.ho_ten)
print("Học sinh có năm sinh từ 2012 đến 2015 và có học lực giỏi:")
for hoc_sinh in danh_sach_hoc_sinh:
    if 2012 <= hoc_sinh.nam_sinh <= 2015 and hoc_sinh.xep_loai() == "Giỏi":
        print(hoc_sinh.ho_ten)
```
