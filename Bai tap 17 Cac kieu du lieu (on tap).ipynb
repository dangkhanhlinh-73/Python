Bài tập 1: Tạo danh sách bằng cách chọn các mục chỉ mục lẻ từ danh sách đầu tiên và các mục chỉ mục chẵn từ danh sách thứ hai

```python
list1 = [1, 3, 5, 7, 9]
list2 = [2, 4, 6, 8, 10]

result = []

# Lặp qua danh sách đầu tiên (list1) và chọn các mục chỉ mục lẻ
for i in range(len(list1)):
    if i % 2 == 0:
        result.append(list1[i])

# Lặp qua danh sách thứ hai (list2) và chọn các mục chỉ mục chẵn
for i in range(len(list2)):
    if i % 2 == 1:
        result.append(list2[i])

print(result)
```

Liệt kê sau khi Thêm phần tử cuối cùng:  [34, 54, 11, 67, 89, 43, 94, 11]
