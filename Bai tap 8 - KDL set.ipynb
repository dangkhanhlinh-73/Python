### Câu 1. (Tổng chữ số chung) Viết CT nhập 2 số tự nhiên a và b. Tính tổng các chữ số chung của a và b. Ví dụ a=12373536, b = 4747373 thì tổng số chung bằng 3+7=10. Chú ý rằng nếu 1 chữ số xuất hiện nhiều lần trong cả a và b thì chữ  số đó chỉ tính 1 lần.

```python
def findcommondigitssum(a, b):
    # Chuyển hai số a và b thành chuỗi để dễ dàng truy cập từng chữ số
    str_a = str(a)
    str_b = str(b)
    
    # Tạo một danh sách để lưu trữ các chữ số chung
    common_digits = []
    
    # Duyệt qua từng chữ số trong a
    for digit in str_a:
        # Kiểm tra xem chữ số này có trong b không và chưa được thêm vào danh sách chữ số chung
        if digit in str_b and digit not in common_digits:
            common_digits.append(digit)
    
    # Tính tổng các chữ số chung
    sumcommondigits = sum(int(digit) for digit in common_digits)
    
    return sumcommondigits

# Nhập hai số tự nhiên a và b
a = int(input("Nhập số tự nhiên a: "))
b = int(input("Nhập số tự nhiên b: "))

# Tính tổng các chữ số chung và in kết quả
result = findcommondigitssum(a, b)
print(f"Tổng các chữ số chung của {a} và {b} là: {result}")
```
