Câu 1. (Tạo từ điển) CHo 2 DS: 1 DS gồm n số khác nhau [a_0, ..., a_n-1] và 1 DS gồm n tên [name_0, ..., name_n-1]. Hãy tạo 1 TĐ mà các phần tử có dạng {a_i : name_i}, sau đó in nội dung TĐ dưới dang: <khóa> : <giá trị>

```python


# Khởi tạo danh sách số và danh sách tên rỗng
numbers = []
names = []

n = int(input("Nhập số lượng phần tử (n): "))

# Nhập danh sách số và tên từ bàn phím
for i in range(n):
    number = int(input(f"Nhập số thứ {i + 1}: "))
    name = input(f"Nhập tên thứ {i + 1}: ")
    numbers.append(number)
    names.append(name)

# Tạo một từ điển từ hai danh sách trên
dictionary = {}

for i in range(n):
    dictionary[numbers[i]] = names[i]

# In nội dung của từ điển dưới dạng khóa : giá trị
for key, value in dictionary.items():
    print(f"{key} : {value}")


















```

Câu 2. (Thống kê điểm thi)  Viết CT nhập thông tin của học sinh gồm: điểm và tên. Các thông tin này lưu trong 1 TĐ. Điểm thuộc {0, 1, ..., 10}. Sau đó thống kê những bạn được điểm 10, điểm 9, ..., điểm 0. 

Nhận thấy: có thể sử dụng điểm làm khóa cho TĐ. Và ứng với mỗi khóa i (i=0,1,...,10) giá trị là 1 DS các tên học sinh có điểm là i. 

```python
# Khởi tạo từ điển để lưu thông tin điểm thi của học sinh
thongkediem = {i: [] for i in range(11)}

# Hàm để nhập thông tin học sinh và thêm vào từ điển
def themhocsinh():
    ten = input("Nhập tên học sinh: ")
    diem = int(input("Nhập điểm của học sinh (từ 0 đến 10): "))
    if diem < 0 or diem > 10:
        print("Điểm không hợp lệ. Vui lòng nhập lại.")
    else:
        thongkediem[diem].append(ten)

# Hàm để thực hiện thống kê và in ra danh sách học sinh theo điểm
def thongkevain():
    for diem, danh_sach in thongkediem.items():
        print(f"Điểm {diem}: {', '.join(danh_sach)}")

# Vòng lặp chính cho phép người dùng thêm thông tin học sinh và thực hiện thống kê
while True:
    print("\n1. Thêm thông tin học sinh")
    print("2. Thống kê và in danh sách")
    print("3. Thoát")
    luachon = input("Chọn một tùy chọn (1/2/3): ")
    
    if luachon == '1':
        themhocsinh()
    elif luachon == '2':
        thongkevain()
    elif luachon == '3':
        break
    else:
        print("Tùy chọn không hợp lệ. Vui lòng chọn lại.")
```
