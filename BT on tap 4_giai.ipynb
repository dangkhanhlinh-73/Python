Bài tập 1: Một Robot di chuyển trong mặt phẳng bắt đầu từ điểm đầu tiên (0,0). Robot có thể di chuyển theo hướng UP, DOWN, LEFT và RIGHT với những bước nhất định. Dấu di chuyển của robot được đánh hiển thị như sau: 

HUONG K

Trong đó: 
HƯỚNG: UP, DOWN, LEFT và RIGHT
K: Số bước đi theo hướng đó

Hãy viết chương trình Python thực hiện các thao tác sau:
1. Tạo 01 text tên là data.txt chứa thông tin các bước di chuyển của Robot

Ví dụ:
DOWN 4
UP 3
RIGHT 2
LEFT 4

2. Đọc file data.txt và lưu vào 01 danh sách Router, in kết ra màn hình.
3. Từ quãng đường Robot đã di chuyển theo danh sách Router, hãy tính khoảng cách từ vị trí hiện tại đến vị trí đầu tiên. 
4. Hãy viết hàm  Dichuyen(n,k) trong đó n là số bước Robot di chuyển. Trong mỗi lần di chuyển, robot sẽ chọn ngẫu nhiên 1 trong 4 hướng: UP, DOWN, LEFT, RIGHT và số bước di chuyển là một số ngẫu nhiên từ 1 đến k.

```python
f=open('data4.txt','r')
r=f.read().split(' ')
print(r)
```

```python
import math
v=[0,0]
for i in range(len(r)):
     if i%2==0:
        if r[i]=='DOWN':
            v[1]-=int(r[i+1])
        if r[i]=='UP':
            v[1]+=int(r[i+1])
        if r[i]=='LEFT':
            v[0]-=int(r[i+1])
        if r[i]=='RIGHT':
            v[0]+=int(r[i+1])  
print(v)
d = math.sqrt(v[0]*v[0]+v[1]*v[1])
print(d)
```

```python
f=open('data4.txt','r')
r=f.read().split(' ')
import random
n=int(input("Nhập số lần di chuyển: "))
k=int(input("Nhập k: "))
def Dichuyen(n,k):
    v=[0,0]
    d=['UP','DOWN','LEFT','RIGHT']
    for i in range(n):
        j=random.randint(1,k) 
        h=random.choice(d)
        if h=='UP': v[1]+=j
        if h=='DOWN': v[1]-=j
        if h=='LEFT': v[0]-=j
        if h=='RIGHT': v[0]+=j
        print(h,j,v)
    return v
v2 = Dichuyen(n,k)
print(v2)
```
