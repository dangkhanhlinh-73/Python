# Python Lambda

Một hàm lambda là một hàm ẩn danh nhỏ.

Một hàm lambda có thể nhận bất kỳ số lượng đối số nào, nhưng chỉ có thể có một biểu thức.

Cú pháp

lambda arguments : expression

Các ví dụ:

Ví dụ 1: Nhân đối số a với đối số b và trả về kết quả:

```python
def nhan(a,b):
    return a*b
print(nhan(5, 6))
```

```python
x = lambda a,b: a*b
print(x(5,6))



```

```python
def ichan(n):
    if n%2 == 0:
        return True
    else:
        return False
```

```python
x = lambda n : n%2==0
print(x(5))
```

Ví dụ 2: Tổng các đối số a, b và c và trả về kết quả:

```python
x = lambda a, b, c : a + b + c
print(x(5, 6, 2))
```

## Tại sao sử dụng các hàm Lambda?

Sức mạnh của lambda được thể hiện rõ hơn khi bạn sử dụng chúng như một hàm ẩn danh bên trong một hàm khác.

Ví dụ: 
Giả sử bạn có một định nghĩa hàm nhận một đối số và đối số đó sẽ được nhân với một số chưa biết:

```python
def myfunc(n):
  return lambda a : a * n
```

Sử dụng định nghĩa hàm đó để tạo một hàm luôn tăng gấp đôi số bạn gửi vào:

```python
def myfunc(n):
  return lambda a : a * n

mydoubler = myfunc(2)

print(mydoubler(11))
```

Hoặc, sử dụng cùng một định nghĩa hàm để tạo một hàm luôn tăng gấp ba lần số bạn gửi:

```python
def myfunc(n):
  return lambda a : a * n

mytripler = myfunc(3)

print(mytripler(11))
```

Hoặc, sử dụng cùng một định nghĩa hàm để tạo cả hai hàm, trong cùng một chương trình:

```python
def myfunc(n):
  return lambda a : a * n

mydoubler = myfunc(2)
mytripler = myfunc(3)

print(mydoubler(11))
print(mytripler(11))
```

## Sử dụng hàm lambda trong sắp xếp (sort) một danh sách, từ điển, tuple

Python - Sắp xếp danh sách

Đối tượng danh sách có phương thức sort () sẽ sắp xếp danh sách theo kiểu chữ và số, tăng dần, theo mặc định:

Ví dụ: sắp xếp theo alphabet

```python
thislist = ["orange", "mango", "kiwi", "pineapple", "banana"]
thislist.sort()
print(thislist)
```

Ví dụ: sắp xếp theo số tăng dần

```python
thislist = [100, 50, 65, 82, 23]
thislist.sort()
print(thislist)
```

Sắp xếp giảm dần:
Để sắp xếp giảm dần, hãy sử dụng đối số từ khóa reverse = True:

```python
thislist = ["orange", "mango", "kiwi", "pineapple", "banana"]
thislist.sort(reverse = True)
print(thislist)
```

```python
thislist = [100, 50, 65, 82, 23]
thislist.sort(reverse = True)
print(thislist)
```

### Tùy chỉnh chức năng sắp xếp

Ta cũng có thể tùy chỉnh chức năng của riêng mình bằng cách sử dụng đối số từ khóa 

key = function.

Hàm sẽ trả về một số sẽ được sử dụng để sắp xếp danh sách (số thấp nhất đầu tiên):

Ví dụ
Sắp xếp danh sách dựa trên mức độ gần của con số với 50:

```python
thislist = [100, 50, 65, 82, 23]
def myfunc(x):
    return abs(x-50)
thislist.sort(key = myfunc)
print(thislist)
```

Bài tập: Sử dụng hàm lambda để thay thế hàm?

```python
thislist = [100, 50, 65, 82, 23]
thislist.sort(key = lambda x: abs(x-50))
print(thislist)
```

### Sử dụng hàm lambda trong lọc (filter) một danh sách, từ điển, tuple

Hàm filter () trả về một trình lặp là các mục được lọc thông qua một hàm để kiểm tra xem mục đó có được chấp nhận hay không.

Cú pháp:
filter(function, iterable)

function: Một hàm được chạy cho từng mục trong tệp có thể lặp lại
iterable: Có thể lặp lại để lọc

Ví dụ: 

```python
ages = [5, 12, 17, 18, 24, 32]

def myFunc(x):
  if x < 18:
    return False
  else:
    return True

adults = filter(myFunc, ages)

for x in adults:
  print(x)
```

Question? : Viết chương trình trên dùng hàm lambda

```python
ages = [5, 12, 17, 18, 24, 32]
adults = list(filter(lambda x : x >= 18, ages))
print(adults)
```
