Câu 3. (Nén chuỗi) Cho 1 chuỗi gồm các ký tự chữ cái latinh thường. Thực hiện nén chuỗi theo quy tắc sau: với 1 dãy các ký tự giống nhau liên tiếp, hãy thay thế dãy này bằng <ký tự><số lần xuất hiện>.
Ví dụ: chuỗi "dddaeeeedda" thay bằng "d3ae4d2a".
Chú ý: không nén "ddd" thành "d2d", và "a" thành "a1".

```python
def nen_chuoi(chuoi):
    if not chuoi:
        return ""

    nen = ""
    ky_tu_hien_tai = chuoi[0]
    dem = 1

    for i in range(1, len(chuoi)):
        if chuoi[i] == ky_tu_hien_tai:
            dem += 1
        else:
            nen += ky_tu_hien_tai + (str(dem) if dem > 1 else "")
            ky_tu_hien_tai = chuoi[i]
            dem = 1

    nen += ky_tu_hien_tai + (str(dem) if dem > 1 else "")
    return nen

# Ví dụ
chuoi = "dddaeeeedda"
chuoi_nen = nen_chuoi(chuoi)
print(chuoi_nen)  # Kết quả: "d3ae4d2a"
```

Câu 4. (Giải nén chuỗi) Cho chuỗi str2 được nén bởi chuỗi str1 theo quy tắc trên (câu 3). Hãy đưa ra chuỗi str1.

```python
def giai_nen_chuoi(str2):
    str1 = ""
    i = 0

    while i < len(str2):
        ky_tu = str2[i]
        i += 1
        so_lan_xuat_hien = ""

        while i < len(str2) and str2[i].isdigit():
            so_lan_xuat_hien += str2[i]
            i += 1

        so_lan_xuat_hien = int(so_lan_xuat_hien) if so_lan_xuat_hien else 1
        str1 += ky_tu * so_lan_xuat_hien

    return str1

# Ví dụ
str2 = "d3ae4d2a"
str1 = giai_nen_chuoi(str2)
print(str1)  # Kết quả: "dddaeeeedda"
```
