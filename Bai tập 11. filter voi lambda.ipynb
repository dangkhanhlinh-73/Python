Bài tập 1: Viết chương trình Python để lọc danh sách các số nguyên theo chẵn, lẽ bằng Lambda

```python

n_str = input("Nhập số nguyên n: ")

try:
    n = int(n_str)
    if n < 1:
        print("Vui lòng nhập một số nguyên dương lớn hơn 0.")
    else:
        # Tạo danh sách từ 1 đến n
        numbers = list(range(1, n + 1))
        
        # Sử dụng lambda để lọc danh sách các số chẵn và lẻ
        even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
        odd_numbers = list(filter(lambda x: x % 2 != 0, numbers))
        
        # In danh sách các số chẵn và lẻ
        print("Các số chẵn từ 1 đến", n, "là:", even_numbers)
        print("Các số lẻ từ 1 đến", n, "là:", odd_numbers)
except ValueError:
    print("Lỗi: Đây không phải là một số nguyên hợp lệ.")

```

Bài tập 2: Viết chương trình Python để lọc chiều cao và chiều rộng của sinh viên, được lưu trữ trong từ điển bằng lambda
Ví dụ Từ điển gốc:
{'Anh': (6.2, 70), 'Đạt': (5.9, 65), 'Thành': (6.0, 68), 'PHUy': (5.8, 66)}
Height> 6ft and Weight> 70kg
Anh': (6.2, 70)

```python

students = {'Anh': (6.2, 70), 'Đạt': (5.9, 65), 'Thành': (6.0, 68), 'PHUy': (5.8, 66)}

# Sử dụng lambda để lọc sinh viên theo điều kiện
filtered_students = dict(filter(lambda item: item[1][0] > 6.0 and item[1][1] > 70, students.items()))

print("Sinh viên có chiều cao > 6ft và cân nặng > 70kg:")
for name, (height, weight) in filtered_students.items():
    print(f"{name}: (Chiều cao: {height}ft, Cân nặng: {weight}kg)")
```
