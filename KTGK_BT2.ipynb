Bài tập 1: Viết các hàm trong Python thực hiện các công việc sau:
1. Nhập vào 01 danh sách các số nguyên từ bàn phím
2. Tính tổng, giá trị trung bình các số trong danh sách nhập từ hàm ở câu 1
3. Tìm các giá trị lớn nhất, bé nhất trong danh sách nhập từ hàm ở câu 1
4. Hoán đổi 02 vị trí trong danh sách: vị trí đầu tiên và vị trí cuối cùng
5. Hoán đổi 02 vị trí trong danh sách: vị trí pos 1 và pos2
6. In các số chẵn trong một danh sách
7. Trả về danh sách các phần tử trong 01 danh sách cho trước mà phần tử đó có số lần xuất hiện trong danh sách lớn hơn một số k cho trước
8. Kiểm tra xem danh sách cho trước có chứa dữ liệu của một dãy các số liên tiếp hay không
9. Viết chương trình để chạy và thực hiện các hàm trên

```python
def nhap_danh_sach():
    danh_sach = list(map(int, input("Nhập danh sách các số nguyên, cách nhau bằng dấu cách: ").split()))
    return danh_sach
def tinh_tong_trung_binh(danh_sach):
    tong = sum(danh_sach)
    trung_binh = tong / len(danh_sach)
    return tong, trung_binh

def tim_max_min(danh_sach):
    max_value = max(danh_sach)
    min_value = min(danh_sach)
    return max_value, min_value
def hoan_doi_dau_cuoi(danh_sach):
    danh_sach[0], danh_sach[-1] = danh_sach[-1], danh_sach[0]
    return danh_sach
def hoan_doi_vi_tri(danh_sach, pos1, pos2):
    danh_sach[pos1], danh_sach[pos2] = danh_sach[pos2], danh_sach[pos1]
    return danh_sach
def in_so_chan(danh_sach):
    so_chan = [x for x in danh_sach if x % 2 == 0]
    print("Các số chẵn trong danh sách:", so_chan)
def phan_tu_xuat_hien_nhieu(danh_sach, k):
    phan_tu_xuat_hien_nhieu = [x for x in danh_sach if danh_sach.count(x) > k]
    return phan_tu_xuat_hien_nhieu
def kiem_tra_lien_tiep(danh_sach):
    danh_sach.sort()
    return all(danh_sach[i] + 1 == danh_sach[i + 1] for i in range(len(danh_sach) - 1))
danh_sach = nhap_danh_sach()
print("Danh sách vừa nhập:", danh_sach)

tong, trung_binh = tinh_tong_trung_binh(danh_sach)
print("Tổng các số trong danh sách:", tong)
print("Giá trị trung bình của các số trong danh sách:", trung_binh)

max_value, min_value = tim_max_min(danh_sach)
print("Giá trị lớn nhất trong danh sách:", max_value)
print("Giá trị bé nhất trong danh sách:", min_value)

danh_sach_hoan_doi_dau_cuoi = hoan_doi_dau_cuoi(danh_sach.copy())
print("Danh sách sau khi hoán đổi vị trí đầu tiên và cuối cùng:", danh_sach_hoan_doi_dau_cuoi)

pos1 = int(input("Nhập vị trí pos1 cần hoán đổi: "))
pos2 = int(input("Nhập vị trí pos2 cần hoán đổi: "))
danh_sach_hoan_doi_vi_tri = hoan_doi_vi_tri(danh_sach.copy(), pos1, pos2)
print(f"Danh sách sau khi hoán đổi vị trí {pos1} và {pos2}:", danh_sach_hoan_doi_vi_tri)

in_so_chan(danh_sach)

k = int(input("Nhập giá trị k: "))
phan_tu_xuat_hien_nhieu_k = phan_tu_xuat_hien_nhieu(danh_sach, k)
print(f"Các phần tử xuất hiện nhiều hơn {k} lần trong danh sách:", phan_tu_xuat_hien_nhieu_k)

if kiem_tra_lien_tiep(danh_sach):
    print("Danh sách chứa dữ liệu của một dãy số liên tiếp.")
else:
    print("Danh sách không chứa dữ liệu của một dãy số liên tiếp.")
```
