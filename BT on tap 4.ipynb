Bài tập 1: Một Robot di chuyển trong mặt phẳng bắt đầu từ điểm đầu tiên (0,0). Robot có thể di chuyển theo hướng UP, DOWN, LEFT và RIGHT với những bước nhất định. Dấu di chuyển của robot được đánh hiển thị như sau: 

HUONG K

Trong đó: 
HƯỚNG: UP, DOWN, LEFT và RIGHT
K: Số bước đi theo hướng đó

Hãy viết chương trình Python thực hiện các thao tác sau:
1. Tạo 01 text tâm là data.txt chứa thông tin các bước di chuyển của Robot

Ví dụ:
DOWN 4
UP 3
RIGHT 2
LEFT 4

2. Đọc file data.txt và lưu vào 01 danh sách Router, in kết ra màn hình.
3. Từ quãng đường Robot đã di chuyển theo danh sách Router, hãy tính khoảng cách từ vị trí hiện tại đến vị trí đầu tiên. 
4. Hãy viết hàm  Dichuyen(n,k) trong đó n là số bước Robot di chuyển. Trong mỗi lần di chuyển, robot sẽ chọn ngẫu nhiên 1 trong 4 hướng: UP, DOWN, LEFT, RIGHT và số bước di chuyển là một số ngẫu nhiên từ 1 đến k.

```python
import random
def tao_file_data1(file_path, so_buoc):
    huongs = ['UP', 'DOWN', 'LEFT', 'RIGHT']
    with open(file_path, 'w') as file:
        for _ in range(so_buoc):
            huong = random.choice(huongs)
            buoc = random.randint(1, 5) 
            file.write(f'{huong} {buoc} '
def doc_file_data1(file_path):
    with open(file_path, 'r') as file:
        du_lieu = file.read().split()
        return [(du_lieu[i], int(du_lieu[i+1])) for i in range(0, len(du_lieu), 2)]
def tinh_khoang_cach(router):
    x, y = 0, 0
    for huong, buoc in router:
        if huong == 'UP':
            y += buoc
        elif huong == 'DOWN':
            y -= buoc
        elif huong == 'LEFT':
            x -= buoc
        elif huong == 'RIGHT':
            x += buoc
    return abs(x) + abs(y)
def di_chuyen(n, k):
    huongs = ['UP', 'DOWN', 'LEFT', 'RIGHT']
    router = []
    for _ in range(n):
        huong = random.choice(huongs)
        buoc = random.randint(1, k)
        router.append((huong, buoc))
    return router

so_buoc = 5
file_path = 'data.txt'
tao_file_data1(file_path, so_buoc)

router = doc_file_data1(file_path)
print('Danh sách Router:', router)
khoang_cach = tinh_khoang_cach(router)
print('Khoảng cách:', khoang_cach)
n = 3  
k = 5 
new_router = di_chuyen(n, k)
print('Danh sách Router mới:', new_router)
```
