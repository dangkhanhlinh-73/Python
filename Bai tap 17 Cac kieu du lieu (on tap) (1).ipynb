Bài tập 1: Tạo danh sách bằng cách chọn các mục chỉ mục lẻ từ danh sách đầu tiên và các mục chỉ mục chẵn từ danh sách thứ hai

```python
list1 = [1, 3, 5, 7, 9]
list2 = [2, 4, 6, 8, 10]

result = []

# Lặp qua danh sách đầu tiên (list1) và chọn các mục chỉ mục lẻ
for i in range(len(list1)):
    if i % 2 == 0:
        result.append(list1[i])

# Lặp qua danh sách thứ hai (list2) và chọn các mục chỉ mục chẵn
for i in range(len(list2)):
    if i % 2 == 1:
        result.append(list2[i])

print(result)

```

Bài tập 3: Cắt danh sách thành 3 phần bằng nhau và đảo ngược từng phần

```python
def cat_va_dao_nguoc_danh_sach(danh_sach):
    # Tính độ dài của danh sách và chia nó thành 3 phần
    do_dai = len(danh_sach)
    phan_du = do_dai % 3
    kich_thuoc_phan = do_dai // 3

    # Khởi tạo danh sách kết quả
    ket_qua = []

    # Tách danh sách thành 3 phần bằng nhau và đảo ngược từng phần
    for i in range(0, do_dai, kich_thuoc_phan):
        phan = danh_sach[i:i + kich_thuoc_phan]
        phan_dao_nguoc = phan[::-1]
        ket_qua.extend(phan_dao_nguoc)

    return ket_qua

# Ví dụ sử dụng
danh_sach = [1, 2, 3, 4, 5, 6, 7, 8, 9]
ket_qua = cat_va_dao_nguoc_danh_sach(danh_sach)
print(ket_qua)
```

Bài tập 4: Đếm số lần xuất hiện của mỗi phần tử trong danh sách
Viết chương trình lặp một danh sách đã cho và đếm số lần xuất hiện của từng phần tử và tạo từ điển để hiển thị số lượng của từng phần tử.

```python
def dem_so_lan_xuat_hien(danh_sach):
    # Tạo một từ điển để lưu trữ số lần xuất hiện của từng phần tử
    dem = {}
    
    # Lặp qua danh sách và đếm số lần xuất hiện của từng phần tử
    for phan_tu in danh_sach:
        if phan_tu in dem:
            dem[phan_tu] += 1
        else:
            dem[phan_tu] = 1
    
    return dem

# Ví dụ sử dụng hàm với một danh sách bất kỳ
danh_sach = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
ket_qua = dem_so_lan_xuat_hien(danh_sach)

# In kết quả
for phan_tu, so_lan_xuat_hien in ket_qua.items():
    print(f"{phan_tu}: {so_lan_xuat_hien} lần")
```

Bài tập 5: Tạo một tập hợp trong Python sao cho nó hiển thị phần tử từ cả hai danh sách trong một cặp

```python
# Nhập danh sách thứ nhất từ người dùng
list1 = input("Nhập danh sách thứ nhất, cách nhau bằng dấu cách: ").split()

# Nhập danh sách thứ hai từ người dùng
list2 = input("Nhập danh sách thứ hai, cách nhau bằng dấu cách: ").split()

# Chuyển danh sách thành tập hợp và lọc các giá trị trùng lặp
set1 = set(list1)
set2 = set(list2)

# Tạo tập hợp chứa các phần tử từ cả hai danh sách
union_set = set1.union(set2)

# In tập hợp kết quả
print("Tập hợp chứa các phần tử từ cả hai danh sách là:", union_set)
```

Bài tập 6: Tìm giao (chung) của hai tập hợp và loại bỏ các phần tử đó khỏi tập hợp thứ nhất

```python
# Nhập tập hợp thứ nhất từ bàn phím
n = int(input("Nhập số phần tử của tập hợp thứ nhất: "))
tap_hop_1 = set()
for i in range(n):
    phan_tu = input(f"Nhập phần tử {i + 1}: ")
    tap_hop_1.add(phan_tu)

# Nhập tập hợp thứ hai từ bàn phím
m = int(input("Nhập số phần tử của tập hợp thứ hai: "))
tap_hop_2 = set()
for i in range(m):
    phan_tu = input(f"Nhập phần tử {i + 1}: ")
    tap_hop_2.add(phan_tu)

# Tìm giao chung của hai tập hợp
giao_chung = tap_hop_1.intersection(tap_hop_2)

# Loại bỏ các phần tử giao chung khỏi tập hợp thứ nhất
tap_hop_1.difference_update(giao_chung)

# In kết quả
print("Giao chung của hai tập hợp là:", giao_chung)
print("Tập hợp thứ nhất sau khi loại bỏ các phần tử giao chung:", tap_hop_1)
```

Bài tập 7: Lặp lại một danh sách đã cho và kiểm tra xem một phần tử đã cho có tồn tại dưới dạng giá trị của khóa trong từ điển hay không. Nếu không, hãy xóa nó khỏi danh sách
Ví dụ: 
roll_number = [47, 64, 69, 37, 76, 83, 95, 97]
sample_dict = {'Jhon':47, 'Emma':69, 'Kelly':76, 'Jason':97}
Kết quả Sau khi xóa các phần tử không mong muốn khỏi danh sách: 
[47, 69, 76, 97]

```python
roll_number = [47, 64, 69, 37, 76, 83, 95, 97]
sample_dict = {'Jhon':47, 'Emma':69, 'Kelly':76, 'Jason':97}

# Tạo một danh sách tạm thời để chứa các phần tử cần giữ lại
temp_list = []

# Lặp qua danh sách roll_number
for num in roll_number:
    # Kiểm tra xem giá trị num có trong từ điển sample_dict không
    if num in sample_dict.values():
        temp_list.append(num)

# Gán danh sách tạm thời là danh sách mới cho roll_number
roll_number = temp_list

print(roll_number)
```

Bài tập 9: Xóa các phần tử trùng lặp ra khỏi danh sách và tạo một bộ (tuple) và tìm số tối thiểu và số tối đa

```python
# Đầu tiên, bạn có thể tạo một danh sách ban đầu
danh_sach = [1, 2, 2, 3, 4, 4, 5]

# Sử dụng set để loại bỏ các phần tử trùng lặp
danh_sach_khong_trung_lap = list(set(danh_sach))

# Sau đó, tạo một bộ tuple
bo_tuple = tuple(danh_sach_khong_trung_lap)

# Tìm số tối thiểu và số tối đa
so_toi_thieu = min(bo_tuple)
so_toi_da = max(bo_tuple)

print("Danh sách ban đầu:", danh_sach)
print("Danh sách sau khi loại bỏ trùng lặp:", danh_sach_khong_trung_lap)
print("Bộ tuple:", bo_tuple)
print("Số tối thiểu:", so_toi_thieu)
print("Số tối đa:", so_toi_da)
```
