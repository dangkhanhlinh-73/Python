```python
def tinh_tong_le_chan(tup):
    tong_le = 0
    tong_chan = 0
    
    for so in tup:
        if so % 2 == 0:
            tong_chan += so
        else:
            tong_le += so
    
    return tong_le, tong_chan

# Nhập vào bộ tuple
n = int(input("Nhập vào số lượng số nguyên dương: "))
lst = []
for i in range(n):
    so = int(input(f"Nhập số nguyên dương thứ {i + 1}: "))
    lst.append(so)

b = tuple(lst)

# Gọi hàm tính tổng số lẻ và tổng số chẵn
tong_le, tong_chan = tinh_tong_le_chan(b)

# In kết quả
print(f"Tổng các số lẻ trong bộ là: {tong_le}")
print(f"Tổng các số chẵn trong bộ là: {tong_chan}")
```
