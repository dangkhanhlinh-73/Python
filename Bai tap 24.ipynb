Bài tập 1: Viết hàm Python để tìm số tờ tiền (Mẫu tờ tiền: 10, 20, 50, 100, 200 và 500 ) trong một số tiền cho trước. Kết quả trả về từ điển
Ví dụ: 800 = 1x500+1x200+1x100 -> {500:1, 200:1, 100:1}

```python
def tinh_so_to_tien(so_tien):
    # Mẫu tờ tiền
    mau_to_tien = [500, 200, 100, 50, 20, 10]

    # Tạo một từ điển để lưu số tờ của mỗi loại tiền
    ket_qua = {}

    # Duyệt qua từng mẫu tờ tiền
    for to_tien in mau_to_tien:
        # Tính số tờ của mỗi loại tiền và cập nhật vào từ điển
        so_to = so_tien // to_tien
        if so_to > 0:
            ket_qua[to_tien] = so_to
            so_tien %= to_tien

    return ket_qua
so_tien_test = 800
ket_qua_test = tinh_so_to_tien(so_tien_test)
print(f"{so_tien_test} =", end=" ")
print(" + ".join([f"{so_to}x{to_tien}" for to_tien, so_to in ket_qua_test.items()]))
print("Kết quả:", ket_qua_test)
```

Bài tập 2:Viết chương trình Python để tính số nợ trong n tháng. Số tiền vay là 100.000 đô la và khoản vay cộng thêm 5% lãi suất của khoản nợ và làm tròn số tiền đó lên 1.000 đối với hàng tháng. 

```python
def tinh_so_no(n_thang):
    # Số tiền vay ban đầu
    so_tien_vay = 100000
    
    # Lãi suất hàng tháng
    lai_suat = 0.05
    
    # Làm tròn số tiền lên đến 1.000 đô la
    def lam_tron_tien(tien):
        return round(tien, -3)
    
    # Tính số nợ cho mỗi tháng
    for thang in range(1, n_thang + 1):
        # Tính số tiền lãi cho tháng này
        tien_lai = so_tien_vay * lai_suat
        
        # Cộng số tiền lãi vào số tiền vay
        so_tien_vay += tien_lai
        
        # Làm tròn số tiền vay lên đến 1.000 đô la
        so_tien_vay = lam_tron_tien(so_tien_vay)
        
        # In thông tin cho mỗi tháng
        print(f'Tháng {thang}: Số nợ = {so_tien_vay:.2f} đô la')
    return so_tien_vay
so_thang = int(input("Nhập số tháng: "))
so_no_cuoi_cung = tinh_so_no(so_thang)
print(f'Số nợ cuối cùng sau {so_thang} tháng là: {so_no_cuoi_cung:.2f} đô la')
```

Bài tập 3: Viết chương trình Python đọc một văn bản (chỉ các ký tự chữ cái và dấu cách.) Và in ra hai từ. Từ đầu tiên là từ xuất hiện thường xuyên nhất trong văn bản. Cái thứ hai là từ có số lượng chữ cái tối đa.

```python
import re
from collections import Counter

def tim_tu_xuat_hien_nhieu_nhat_va_tu_dai_nhat(van_ban):
    # Loại bỏ các ký tự không phải là chữ cái và dấu cách
    van_ban_da_loc = re.sub(r'[^a-zA-Z\s]', '', van_ban)
    cac_tu = van_ban_da_loc.split()

    tu_xuat_hien_nhieu_nhat = Counter(cac_tu).most_common(1)[0][0]
    
    # Tìm từ có số lượng chữ cái tối đa
    tu_dai_nhat = max(cac_tu, key=len)
    
    return tu_xuat_hien_nhieu_nhat, tu_dai_nhat
van_ban_mau = "Đây là một đoạn văn bản mẫu. Từ xuất hiện thường xuyên nhất ở đây là 'một', và từ dài nhất là 'văn bản mẫu'."

tu_xuat_hien_nhieu_nhat, tu_dai_nhat = tim_tu_xuat_hien_nhieu_nhat_va_tu_dai_nhat(van_ban_mau)

print(f"Từ xuất hiện thường xuyên nhất: {tu_xuat_hien_nhieu_nhat}")
print(f"Từ dài nhất: {tu_dai_nhat}")
```
