# Chuỗi các bài tập về từ điển Birthday

Bài tập 1. Tạo một từ điển (trong file python) tên và ngày sinh. Khi chạy chương trình của mình, nó sẽ yêu cầu người dùng nhập tên và trả lại ngày sinh của người đó cho họ. Sự tương tác sẽ giống như thế này:
>>>Chào mừng đến với birthday dictionary. Chúng tôi biết ngày sinh nhạt của :
Albert Einstein
Benjamin Franklin
Nguyen Tat Thanh
>>> Sinh nhật của người mà bạn muốn tìm?
Benjamin Franklin
>>> Ngày sinh của Benjamin Franklin's là 01/17/1706.

```python
import json

def load_birthday_dictionary():
    try:
        with open("birthday_dict.json", "r") as file:
            birthday_dict = json.load(file)
    except (FileNotFoundError, json.JSONDecodeError):
        birthday_dict = {}
    return birthday_dict

def save_birthday_dictionary(birthday_dict):
    with open("birthday_dict.json", "w") as file:
        json.dump(birthday_dict, file, ensure_ascii=False)

def get_birthday(name, birthday_dict):
    return birthday_dict.get(name, "Không tìm thấy ngày sinh của {}".format(name))

def main():
    birthday_dict = load_birthday_dictionary()

    print("Chào mừng đến với từ điển sinh nhật.")
    print("Chúng tôi biết ngày sinh nhật của:", ", ".join(birthday_dict.keys()))

    while True:
        name = input("Ngày sinh của người mà bạn muốn tìm? ")
        if name.lower() == "exit":
            break
        birthday = get_birthday(name, birthday_dict)
        print("Ngày sinh của {} là: {}".format(name, birthday))

    save_birthday_dictionary(birthday_dict)

if __name__ == "__main__":
    main()
```

Bài tập 2. Trong bài tập này, hãy sửa đổi chương trình của  từ Bài tập 1 để tải từ điển sinh nhật từ tệp trên đĩa (.txt hoặc JSON), thay vì định nghĩa từ điển trong chương trình.

```python
import json

def load_birthday_dictionary(file_path):
    try:
        with open(file_path, "r") as file:
            birthday_dict = json.load(file)
    except (FileNotFoundError, json.JSONDecodeError):
        birthday_dict = {}
    return birthday_dict

def save_birthday_dictionary(birthday_dict, file_path):
    with open(file_path, "w") as file:
        json.dump(birthday_dict, file, ensure_ascii=False)

def get_birthday(name, birthday_dict):
    return birthday_dict.get(name, "Không tìm thấy ngày sinh của {}".format(name))

def main():
    file_path = "birthday_dict.json"  

    birthday_dict = load_birthday_dictionary(file_path)

    print("Chào mừng đến với từ điển sinh nhật.")
    print("Chúng tôi biết ngày sinh nhật của:", ", ".join(birthday_dict.keys()))

    while True:
        name = input("Ngày sinh của người mà bạn muốn tìm? (Nhập 'exit' để kết thúc) ")
        if name.lower() == "exit":
            break
        birthday = get_birthday(name, birthday_dict)
        print("Ngày sinh của {} là: {}".format(name, birthday))

    save_birthday_dictionary(birthday_dict, file_path)

if __name__ == "__main__":
    main()
```

Bài tập 3. Trong bài tập trước, chúng ta đã lưu thông tin về tên và ngày sinh của các người nổi tiếng vào đĩa. Trong bài tập này, hãy tải tệp JSON đó từ đĩa, trích xuất các tháng có sinh nhật và đếm xem có bao nhiêu người có sinh nhật trong mỗi tháng.

```python
import json
from collections import Counter

def load_birthday_dictionary(file_path):
    try:
        with open(file_path, "r") as file:
            birthday_dict = json.load(file)
    except (FileNotFoundError, json.JSONDecodeError):
        birthday_dict = {}
    return birthday_dict

def count_birthdays_by_month(birthday_dict):
    birth_month_counter = Counter()

    for name, birthday in birthday_dict.items():
        birth_month = birthday.split("/")[0]  # Lấy tháng từ ngày sinh
        birth_month_counter[birth_month] += 1

    return birth_month_counter

def main():
    file_path = "birthday_dict.json"  

    birthday_dict = load_birthday_dictionary(file_path)

    print("Chào mừng đến với từ điển sinh nhật.")
    print("Chúng tôi biết ngày sinh nhật của:", ", ".join(birthday_dict.keys()))

    birth_month_counter = count_birthdays_by_month(birthday_dict)

    print("\nSố lượng người có sinh nhật trong mỗi tháng:")
    for month, count in sorted(birth_month_counter.items()):
        print("Tháng {}: {}".format(month, count))

if __name__ == "__main__":
    main()
```
