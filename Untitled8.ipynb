```python
import math

class Iris:
    def __init__(self, length_sepal=None, width_sepal=None, length_petal=None, width_petal=None, flower_type=None):
        self.length_sepal = length_sepal
        self.width_sepal = width_sepal
        self.length_petal = length_petal
        self.width_petal = width_petal
        self.flower_type = flower_type

    def set_flower_type(self, flower_type):
        self.flower_type = flower_type

    def calculate_distance(self, other_flower):
        if not all([self.length_sepal, self.width_sepal, self.length_petal, self.width_petal]) or \
           not all([other_flower.length_sepal, other_flower.width_sepal, other_flower.length_petal, other_flower.width_petal]):
            raise ValueError("Incomplete flower data")

        distance = math.sqrt(
            (self.length_sepal - other_flower.length_sepal) ** 2 +
            (self.width_sepal - other_flower.width_sepal) ** 2 +
            (self.length_petal - other_flower.length_petal) ** 2 +
            (self.width_petal - other_flower.width_petal) ** 2
        )

        return round(distance, 2)

# Ví dụ sử dụng lớp Iris:
iris1 = Iris(length_sepal=5.1, width_sepal=3.5, length_petal=1.4, width_petal=0.2, flower_type="Iris-setosa")
iris2 = Iris(length_sepal=6.3, width_sepal=3.3, length_petal=4.7, width_petal=1.6, flower_type="Iris-versicolour")

# In thông tin của bông hoa
print("Iris 1:", iris1.__dict__)
print("Iris 2:", iris2.__dict__)

# Gán giá trị mới cho kiểu hoa
iris1.set_flower_type("Iris-virginica")
print("Iris 1 (updated):", iris1.__dict__)

# Tính khoảng cách giữa hai bông hoa
distance = iris1.calculate_distance(iris2)
print("Distance between iris1 and iris2:", distance)
```
