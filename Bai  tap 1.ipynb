Câu 1. Cho đường tròn có bán kính R. Tính chu vi của đt đó, làm tròn đến 2 chữ số thập phân.

```python
import math

R=float(input("Nhập vào bán kính R: "))
pi=math.pi
chuvi=2*pi*R
print("Chu vi đường tròn là: ",round(chuvi,2))
```

Câu 2. NHập vào 3 số nguyên dương a,b,c. Tìm UCLN (Greatest Common Divisor) của 3 số đó.


```python
from math import gcd

a=int(input("Nhập vào số nguyên dương a: "))
b=int(input("Nhập vào số nguyên dương b: "))
c=int(input("Nhập vào số nguyên dương c: "))
d=gcd(a,b)

print("UCLN của 3 số a, b, c là: ",gcd(d,c))
```

Xây dựng hàm tính ước số chung lớn nhất USCLN(a,b)

```python
def USCLN(a,b):
    if (a>b):
        if (b == 0): 
            return a
        else:
            return USCLN(b,a%b)
    else:
        if (a == 0): 
            return b
        else:
            return USCLN(a,b%a)
```

```python
a=15
b=10
USCLN(a,b)
```

Câu 3. Nhập vào 2 số nguyên dương a,b. Tìm BCNN của 2 số đó.

```python
from math import gcd

a=int(input("Nhập vào số nguyên dương a: "))
b=int(input("Nhập vào số nguyên dương b: "))

print("BCNN của 2 số a, b là: ",a*b//gcd(a,b))
```
