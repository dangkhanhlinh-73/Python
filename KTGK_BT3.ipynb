Bài tập 1: Viết tập lệnh Python để in một từ điển trong đó các khóa là các số từ 1 đến 20 (bao gồm cả hai) và các giá trị là căn bậc hai của các khóa.

```python
import math
my_dict = {}
for i in range(1, 21):
    my_dict[i] = math.sqrt(i)
for key, value in my_dict.items():
    print(f"{key}: {value}")
```
