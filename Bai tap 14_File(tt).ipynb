Bài tập 1: Ghi tất cả nội dung của một tệp nhất định vào một tệp mới bằng cách bỏ qua dòng số 5 và dòng số 10

Gợi ý: 
Đọc tất cả các dòng từ tệp test.txt bằng phương thức readlines (). Phương thức này trả về tất cả các dòng từ tệp dưới dạng danh sách
Mở tệp văn bản mới ở chế độ ghi (w)
Đặt bộ đếm = 0
Lặp lại từng dòng từ một danh sách
nếu bộ đếm là 4, hãy bỏ qua dòng đó, nếu không hãy ghi dòng đó vào tệp văn bản mới bằng phương thức write ()
Bộ đếm tăng lên 1 trong mỗi lần lặp

Given test.txt file: 
line1
line2
line3
line4
line5
line6
line7

Expected Output: new_file.txt
line1
line2
line3
line4
line6
line7   

```python
# Tên tệp gốc
input_file_name = "test.txt"

# Tên tệp mới
output_file_name = "new_file.txt"

# Mở tệp gốc để đọc
with open(input_file_name, "r") as input_file:
    # Đọc tất cả các dòng từ tệp gốc
    lines = input_file.readlines()

# Mở tệp mới để ghi
with open(output_file_name, "w") as output_file:
    count = 0
    for line in lines:
        count += 1
        if count == 5 or count == 10:
            continue  # Bỏ qua dòng số 5 và dòng số 10
        output_file.write(line)
```

Bài tập 2: Đọc dòng số 4 từ 01 tệp test.txt cho trước

test.txt file:
line1
line2
line3
line4
line5
line6
line7   

```python
# Mở tệp test.txt để đọc
with open('test.txt', 'r') as file:
    # Đọc tất cả các dòng vào một danh sách
    lines = file.readlines()

# Kiểm tra xem tệp có ít nhất 4 dòng không
if len(lines) >= 4:
    # Lấy dòng thứ 4 (số dòng 3 trong danh sách, vì danh sách bắt đầu từ 0)
    line4 = lines[3]
    print(line4)
else:
    print("Tệp không đủ số dòng để đọc dòng thứ 4.")
```
