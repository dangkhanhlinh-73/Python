```python
def get_last_word(input_string):
    words = input_string.split()
    if words:
        return words[-1]
    else:
        return None

def main():
    students = []
    while True:
        student_name = input("Nhập họ tên sinh viên (nhập rỗng để kết thúc): ")
        if not student_name:
            break
        students.append(student_name)
    for student in students:
        last_word = get_last_word(student)
        print(f'Từ cuối cùng của "{student}" là: {last_word}')
    count_anh = sum(1 for student in students if get_last_word(student) == "Anh")
    print(f'\nSố lượng sinh viên có tên là "Anh" là: {count_anh}')

if __name__ == "__main__":
    main()
```

```python
def nhap_danh_sach_hoc_phan():
    danh_sach_hoc_phan = []
    n = int(input("Nhập số lượng học phần: "))
    
    for i in range(n):
        mahp = input("Nhập mã học phần: ")
        tenhp = input("Nhập tên học phần: ")
        sotc = int(input("Nhập số tín chỉ: "))
        
        hoc_phan = {"mahp": mahp, "tenhp": tenhp, "sotc": sotc}
        danh_sach_hoc_phan.append(hoc_phan)
    
    return danh_sach_hoc_phan

def in_thong_tin_nganh_CNTT(danh_sach_hoc_phan):
    print("Các học phần thuộc ngành CNTT:")
    for hoc_phan in danh_sach_hoc_phan:
        if hoc_phan["mahp"][:3] == "TIN":
            print(f"Mã HP: {hoc_phan['mahp']}, Tên HP: {hoc_phan['tenhp']}, Số TC: {hoc_phan['sotc']}")

def in_thong_tin_theo_chuoi(danh_sach_hoc_phan, chuoi):
    print(f"Các học phần có tên chứa '{chuoi}':")
    for hoc_phan in danh_sach_hoc_phan:
        if chuoi.lower() in hoc_phan["tenhp"].lower():
            print(f"Mã HP: {hoc_phan['mahp']}, Tên HP: {hoc_phan['tenhp']}, Số TC: {hoc_phan['sotc']}")

# Thực hiện chương trình
danh_sach_hoc_phan = nhap_danh_sach_hoc_phan()
in_thong_tin_nganh_CNTT(danh_sach_hoc_phan)

chuoi_tim_kiem = input("Nhập chuỗi tìm kiếm: ")
in_thong_tin_theo_chuoi(danh_sach_hoc_phan, chuoi_tim_kiem)
```

```python
3.Xây dựng lớp HOADON mô tả thông tin hóa đơn sử dụng điện của khách hàng,
bao gồm các thuộc tính sau: mã khách hàng, họ tên, địa chỉ, chỉ số cũ, chỉ số mới. Viết
các phương thức cho lớp HOADON bao gồm:
- Phương thức khởi tạo;
- Phương thức trả về mã khách hàng;
- Phương thức trả về số lượng điện sử dụng theo công thức: lượng điện tiêu thụ=
chỉ số mới- chỉ số cũ.
- Phương thức hiển thị thông tin của các thuộc tính mã khách hàng, họ tên, địa chỉ,
số lượng điện tiêu thụ.
- Toán tử tải bội > để so sánh hai đối tượng lớp HOADON dựa vào tiêu chí lượng
điện tiêu thụ.
```

```python
class HOADON:
    def __init__(self, ma_khach_hang, ho_ten, dia_chi, chi_so_cu, chi_so_moi):
        self.ma_khach_hang = ma_khach_hang
        self.ho_ten = ho_ten
        self.dia_chi = dia_chi
        self.chi_so_cu = chi_so_cu
        self.chi_so_moi = chi_so_moi

    def lay_ma_khach_hang(self):
        return self.ma_khach_hang

    def lay_so_luong_dien_tieu_thu(self):
        return self.chi_so_moi - self.chi_so_cu

    def hien_thi_thong_tin(self):
        print(f"Mã khách hàng: {self.ma_khach_hang}")
        print(f"Họ tên: {self.ho_ten}")
        print(f"Địa chỉ: {self.dia_chi}")
        print(f"Số lượng điện tiêu thụ: {self.lay_so_luong_dien_tieu_thu()}")

    def __gt__(self, other):
        return self.lay_so_luong_dien_tieu_thu() > other.lay_so_luong_dien_tieu_thu()
hoadon1 = HOADON("KH001", "Nguyen Van A", "123 ABC Street", 100, 150)
hoadon2 = HOADON("KH002", "Tran Thi B", "456 XYZ Street", 120, 180)

hoadon1.hien_thi_thong_tin()
print("\n")
hoadon2.hien_thi_thong_tin()

if hoadon1 > hoadon2:
    print("\nHóa đơn 1 có lượng điện tiêu thụ lớn hơn hóa đơn 2.")
else:
    print("\nHóa đơn 2 có lượng điện tiêu thụ lớn hơn hóa đơn 1.")
```

```python
class HOADON:
    def __init__(self, ma_khach_hang, ho_ten, dia_chi, so_luong_dien_tieu_thu):
        self.ma_khach_hang = ma_khach_hang
        self.ho_ten = ho_ten
        self.dia_chi = dia_chi
        self.so_luong_dien_tieu_thu = so_luong_dien_tieu_thu
def sort_hoadon_list(hoadon_list):
    return sorted(hoadon_list, key=lambda hoadon: hoadon.so_luong_dien_tieu_thu)
def find_customer_by_ma(ma, hoadon_list):
    for hoadon in hoadon_list:
        if hoadon.ma_khach_hang == ma:
            return hoadon
    return None

# Hàm để lưu thông tin khách hàng có số lượng điện tiêu thụ >100 vào file văn bản
def save_high_consumption_customers(hoadon_list, file_path):
    with open(file_path, 'w') as file:
        for hoadon in hoadon_list:
            if hoadon.so_luong_dien_tieu_thu > 100:
                file.write(f"{hoadon.ma_khach_hang};{hoadon.ho_ten};{hoadon.dia_chi}\n")
if __name__ == "__main__":
    n = int(input("Nhập số lượng đối tượng: "))
    hoadon_list = []

    for i in range(n):
        ma_khach_hang = input("Nhập mã khách hàng: ")
        ho_ten = input("Nhập họ tên: ")
        dia_chi = input("Nhập địa chỉ: ")
        so_luong_dien_tieu_thu = int(input("Nhập số lượng điện tiêu thụ: "))

        hoadon = HOADON(ma_khach_hang, ho_ten, dia_chi, so_luong_dien_tieu_thu)
        hoadon_list.append(hoadon)
    hoadon_list = sort_hoadon_list(hoadon_list)
    ma_to_search = input("Nhập chuỗi ma để tìm kiếm: ")
    found_customer = find_customer_by_ma(ma_to_search, hoadon_list)

    if found_customer:
        print(f"Thông tin khách hàng:\nMã: {found_customer.ma_khach_hang}\nHọ tên: {found_customer.ho_ten}\nĐịa chỉ: {found_customer.dia_chi}")
    else:
        print("Không tồn tại khách hàng này")
    save_high_consumption_customers(hoadon_list, "high_consumption_customers.txt")
```
