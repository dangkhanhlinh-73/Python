Bài tập 1: Viết chương trình Python thực hiện các thao tác sau:
1. Nhập 2 số nguyên dương m, n từ bàn phím. 
2. Tạo ra một mảng 2 chiều A kích thước mxn trong đó giá trị phần tử trong hàng thứ i và cột thứ j của mảng phải là i+j+i*j; i=0,1,...,m-1; j=0,1,...,n-1
3. Tính tổng của các phần tử theo hàng, theo cột, toàn bộ giá trị trong mảng A.
4. Tạo ra một mảng 2 chiều B kích thước mxn trong đó giá trị các phần tử trong mảng là các số thực được chọn ngẫu nhiên từ -10 đến 10
5. Tìm giá trị lớn nhất của các phần tử theo hàng, theo cột ở trong mảng B
6. Tạo mảng C trong đó mỗi phần tử của mảng C là tổng của các phần tử tương ứng trong 2 mảng A, B.
7. Tạo mảng D trong đó:
- Phần tử ở hàng lẻ bằng hiệu của các phần tử tương ứng trong 2 mảng A, B
- Phần tử ở hàng chẵn bằng tích của các phần tử tương ứng trong 2 mảng A, B 

```python
import numpy as np
m = int(input("Nhập giá trị cho m: "))
n = int(input("Nhập giá trị cho n: "))

A = np.array([[i + j + i * j for j in range(n)] for i in range(m)])
tong_theo_hang_A = np.sum(A, axis=1)
tong_theo_cot_A = np.sum(A, axis=0)
tong_toan_bo_A = np.sum(A)
B = np.random.uniform(-10, 10, (m, n))
max_theo_hang_B = np.max(B, axis=1)
max_theo_cot_B = np.max(B, axis=0)
max_toan_bo_B = np.max(B)

C = A + B
D = np.zeros_like(A, dtype=float)
D[1::2, :] = A[1::2, :] - B[1::2, :]
D[::2, :] = A[::2, :] * B[::2, :]
print("Mảng A:")
print(A)

print("\nTổng theo hàng của A:")
print(tong_theo_hang_A)

print("\nTổng theo cột của A:")
print(tong_theo_cot_A)

print("\nTổng toàn bộ giá trị của A:")
print(tong_toan_bo_A)

print("\nMảng B:")
print(B)

print("\nGiá trị lớn nhất theo hàng của B:")
print(max_theo_hang_B)

print("\nGiá trị lớn nhất theo cột của B:")
print(max_theo_cot_B)

print("\nGiá trị lớn nhất toàn bộ của B:")
print(max_toan_bo_B)

print("\nMảng C (tổng của A và B):")
print(C)

print("\nMảng D:")
print(D)
```
