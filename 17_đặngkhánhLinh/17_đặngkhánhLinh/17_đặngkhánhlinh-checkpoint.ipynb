```python
def tinh_tong_diem(diem_qtht, diem_thi):

    tong_diem = {}

    for ma_sv in diem_qtht:

        diem_qtht_sv = diem_qtht.get(ma_sv, 0)

        diem_thi_sv = diem_thi.get(ma_sv, 0)

 

        tong_diem_sv = diem_qtht_sv * 0.4 + diem_thi_sv * 0.6

        tong_diem[ma_sv] = tong_diem_sv


    return tong_diem

diem_qtht = {}

n_qtht = int(input("Nhập số lượng sinh viên có điểm Quá trình học tập: "))

for i in range(n_qtht):

    ma_sv = input("Nhập mã sinh viên: ")

    diem = float(input("Nhập điểm Quá trình học tập: "))

    diem_qtht[ma_sv] = diem

diem_thi = {}

n_thi = int(input("Nhập số lượng sinh viên có điểm Thi: "))

for i in range(n_thi):

    ma_sv = input("Nhập mã sinh viên: ")

    diem = float(input("Nhập điểm Thi: "))

    diem_thi[ma_sv] = diem

tong_diem_sv = tinh_tong_diem(diem_qtht, diem_thi)

print("Tổng điểm của các sinh viên:")

for ma_sv, tong_diem_sv in tong_diem_sv.items():

    print(f"Mã sinh viên {ma_sv}: {tong_diem_sv}")

 
```

```python
def Chuanhoa(X, min_val=0, max_val=1):

 
    minX = min(X)


    maxX = max(X)
 

    if min_val >= max_val:

        raise ValueError("min phải nhỏ hơn max")
 
    new_X = [(x - minX) / (maxX - minX) * (max_val - min_val) + min_val for x in X]

    return new_X
 

X_input = [6, 3, 9, 8, 7]

 

 

 

min_val_input = -1

 

 

 

max_val_input = 1

result = Chuanhoa(X_input, min_val_input, max_val_input)

print(result)
```

```python
import re

 
def preprocess_text(input_file, output_file):

    try:

        with open(input_file, 'r', encoding='utf-8') as file:

            data = file.read()

 

        # Tách câu và xóa ký tự đặc biệt

        sentences = re.split(r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?)\s', data)

        cleaned_sentences = [' '.join(re.findall(r'\b\w+\b', sentence.lower())) for sentence in sentences]

        with open(output_file, 'w', encoding='utf-8') as file:

            for sentence in cleaned_sentences:

                file.write(sentence + '\n')

 

        print(f"Chương trình đã xử lý xong. Kết quả đã được ghi vào file {output_file}")

    except Exception as e:

        print(f"Lỗi: {str(e)}")

 

# Gọi hàm với file dữ liệu và file output

 

preprocess_text("data.txt", "preprocessed_data.txt")

 
```
